## Visualizing Dates

```python
# Extract the year and month from the 'date_added' column
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Plot a countplot for the 'year_added' column
sns.countplot(x='year_added', data=df)
plt.title('Number of Shows Added Each Year')
plt.show()

# Plot a countplot for the 'month_added' column
sns.countplot(x='month_added', data=df)
plt.title('Number of Shows Added Each Month')
plt.show()
```

```python
# Plot histograms for numeric columns
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns].hist(bins=20, figsize=(12, 8))
plt.suptitle('Distribution of Numerical Features')
plt.show()
```

## Visualizing different categorical features

#### Here we can say that our movie data has almost twice the data of TV shows in our dataset 

```python
# Top 10 countries where more movies are released.
top10_countries = df['country'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 5))
top10_countries_plot = top10_countries.plot(kind='bar', rot=45)
plt.title('Top 10 Countries by Count with More Movie Releases')
plt.xlabel('Country')
plt.ylabel('Count')

# Adding labels over the bars with a 45-degree angle
for idx, value in enumerate(top10_countries):
    top10_countries_plot.text(idx, value + 0.1, str(value), ha='center', va='bottom',fontsize=8)

plt.show()
```

### We have significantly more data from United States

```python
# Exploring 'listed_in' column containing genres
# Split genres and explode into separate rows
genres = df['listed_in'].str.split(', ', expand=True).stack()

top_genres = genres.value_counts().head(10)

# Displaying the top 10 genres
print("Top 10 Genres:")
print(top_genres)
```

```python
# Plotting
plt.figure(figsize=(10, 5))
top_genres_plot = top_genres.plot(kind='bar', rot=45)
plt.title('Top 10 Genres by Count')
plt.xlabel('Genre')
plt.ylabel('Count')

# Adding labels over the bars with a 45-degree angle
for idx, value in enumerate(top_genres):
    top_genres_plot.text(idx, value + 0.1, str(value), ha='center', va='bottom', fontsize=8)

plt.show()
```

### International Movies, Dramas, Comedies forms the major chunk of data in the genre category

## Trying to go deeper analyzing the trends for last 2 decades

```python
# Selecting relevant columns
trend_data = df[['type', 'release_year']]

# Filtering for the last two decades (2000 and later)
last_two_decades_trend = trend_data[trend_data['release_year'] >= 2000]

# Group by release year and count the number of each content type
trend_df = last_two_decades_trend.groupby(['release_year', 'type']).size().reset_index(name='Total Content')

# Plotting the trend of each content type over the years
plt.figure(figsize=(12, 8))
trend_plot = sns.lineplot(x='release_year', y='Total Content', data=trend_df, hue='type', marker='o', palette='viridis')
trend_plot.set(title='Trend of Each Type of Content Released Over the Last Two Decades', xlabel='Release Year', ylabel='Total Content')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Content Type', loc='upper left')
plt.show()
```

### Altering the rating feature in the dataset to make it more readable by implementing the mapping function

```python
distinct_ratings = df['rating'].unique()
print("Distinct Ratings:")
print(distinct_ratings)
```

```python
# Mapping for more readable ratings
readable_ratings_mapping = {
    'PG-13': 'Parental Guidance 13',
    'TV-MA': 'TV Mature Audiences',
    'TV-PG': 'TV Parental Guidance',
    'TV-14': 'TV 14 and older',
    'TV-Y7': 'TV Youth 7 and older',
    'TV-Y': 'TV Youth',
    'PG': 'Parental Guidance',
    'TV-G': 'TV General Audiences',
    'R': 'Restricted',
    'G': 'General Audiences',
    'NC-17': 'No one 17 and under admitted',
    'NR': 'Not Rated',
    'TV-Y7-FV': 'TV Youth 7 and older (Fantasy Violence)',
    'UR': 'Unrated'
}

# Replace existing ratings with more readable ones
df['rating'] = df['rating'].map(readable_ratings_mapping)

distinct_ratings = df['rating'].unique()
print("Distinct Ratings:")
print(distinct_ratings)

```

```python

plt.figure(figsize=(10, 6))

# Plotting the count of Top 5 Rating by Movies and Shows
sns.countplot(x="rating", data=df, order=df['rating'].value_counts().head(5).index)

plt.title("Count of Ratings by Movies and Shows")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

### Here we can conclude that 'TV Mature Audiences' and 'TV 14 and older' forms the major chunk of data in the 'ratings' feature
